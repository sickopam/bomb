```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

ds = pd.read_csv("./data/cleaned_sneaks.csv")

display(ds.head())

label_enc = LabelEncoder()
ds["brand_enc"] = label_enc.fit_transform(ds["brand"])

scaler = MinMaxScaler()
num_col = ["retail", "lowestAsk", "highestBid", "annualHigh", "annualLow", "volatility", "deadstockSold", "pricePremium", "averageDeadstockPrice", "lastSale"]

ds[num_col] = scaler.fit_transform(ds[num_col])

display(ds.head())
```

```python
ds.to_csv("pp_sneaks.csv", index=False)
```
