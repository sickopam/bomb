```python
import pandas as pd
import os

ds = pd.read_csv("./data/clean_sneaks.csv")
ds.head()
```

```python
foldername = "./pics"
folder = set(os.listdir(foldername))

def update_path(row):
    shoe_name = row["item"]
    if shoe_name in folder:
        return f"pics/{shoe_name}"
    return row["img_path"]

ds["img_path"] = ds.apply(update_path, axis=1)

ds.head()
```

```python
ds.to_csv("clean_sneaks_update.csv", index=False)
```
